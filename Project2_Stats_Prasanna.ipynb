```python
from scipy.stats import t
t_critical=t.ppf(0.95,29)
#alpha =0.95 and degree of freedom is n-1 i.e 30-1=29
```

```python
t_critical
#finding t_critical_value
```

```python
import numpy as np
sd=np.std(df['Diff'],ddof=1)
#finding standard deviation to derive formal of t_Stat=x_diff_mean-0/standard error
```

```python
se=np.sqrt(30)
#finding standard error
```

```python
x_diff_mean=t_critical*se
```

```python
x_diff_mean
```

```python
t_stat=(x_diff_mean-5)/se
#At diff of 5 , finding t_stat value
#Assumption is that, at the mean difff of 5 p_value wil raise 
```

```python
t_stat
```

```python
p_val=t.cdf(t_stat,29)
#Finding p_value for t_stat 
```

```python
p_val
#this is beta
#Answer for 4 -3 is 0.64
```
